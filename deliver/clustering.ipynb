```python
import cv2 
import numpy as np
from matplotlib import pyplot as plt
from sklearn.cluster import KMeans
```

```python
imag  = cv2.imread('../data/samples/3pe4g.png',0)
```

```python
threshold = 100
ret, im_bin = cv2.threshold(imag,threshold, 255, cv2.THRESH_BINARY_INV)
imag
```

```python
ind = np.nonzero(im_bin)
plt.scatter(ind[1],np.flip(ind[0]))
# plt.plot(ind)
# plt.plot(im_bin)
# cv2.imshow('im',im_bin)
```

```python
X = np.vstack((ind[1],np.flip(ind[0]))).T
X.shape
```

```python
X
```

```python
l = []
thres = 3
for i in range(200):
    if len(X[X[:,0] == i]) > thres:
        for e in X[X[:,0] == i]:
            l.append(e)
            print(e)
```

```python
X = np.array(l)
```

```python
kmeans = KMeans(n_clusters=5, init=np.array([(i*200/6.0, 25) for i in range(1,6)]))
kmeans.fit(X)
y_kmeans = kmeans.predict(X)
```

```python
np.array([(i*200/6.0, 25) for i in range(1,6)]).shape
```

```python

plt.scatter(X[:, 0], X[:, 1], c=y_kmeans, s=50, cmap='viridis')

centers = kmeans.cluster_centers_
plt.scatter(centers[:, 0], centers[:, 1], c='black', s=200, alpha=0.5);
```

```python
X_l = X[:,0]
X_l = X_l.reshape(-1,1)
```

```python
kmeans = KMeans(n_clusters=5, init=np.array([i*200/6.0 for i in range(1,6)]).reshape(-1,1))
kmeans.fit(X_l)
y_kmeans = kmeans.predict(X_l)
```

```python
plt.scatter(X[:, 0], X[:, 1], c=y_kmeans, s=50, cmap='viridis')

centers = kmeans.cluster_centers_
plt.scatter(centers[:], np.ones(centers.shape[0])*25, c='black', s=200, alpha=0.5);
```
